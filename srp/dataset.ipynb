```python
additional2_df = additional2_df[['type', 'product', 'Standard reduction potential (V)', '비고 (용매는 water)']]
additional2_df['solvent'] = 'H2O'
additional2_df.rename(columns={'product' : 'SMILE', 'Standard reduction potential (V)': 'SRP', '비고 (용매는 water)' : 'ph'}, inplace=True)
```

```python
def classify_type(value):
    if pd.isna(value):
        return value  # NaN 값은 건너뛰기
    else:
        check = value.split('→')[1][-1]
        if check == '−':
            return 'anion'
        else:
            return 'neutral'
    

additional2_df['SMILE'] = additional2_df['SMILE'].fillna(method='ffill')
additional2_df['type'] = additional2_df['type'].apply(classify_type)
additional2_df['type'] = additional2_df['type'].fillna(method='ffill')
```

```python
additional2_df = additional2_df[~additional2_df['ph'].astype(str).str.contains('계산')]
```

```python
additional2_df['ph'].unique()
```

```python
condition = additional2_df['ph'].apply(lambda x: x == 'solvent: iso-propyl alcohol')

# 조건에 맞는 인덱스 가져오기
matching_indices = additional2_df[condition].index.tolist()
```

```python
matching_indices
```

```python
additional2_df.loc[matching_indices, 'solvent'] = 'iso-propyl alcohol'
```

```python
additional2_df['solvent'].unique()
```

```python

def classify_ph(value):
    if pd.isna(value):
        return 7
    elif value in ["midpoint potential", "solvent: iso-propyl alcohol", "ref가 여러개일 경우 제일 최근에 나온 논문에서 나온 값을 선택"]:
        return 7
    elif isinstance(value, str) and 'pH' in value:
        val = value.split()[-1]
        return float(val)

additional2_df['ph'] = additional2_df['ph'].apply(classify_ph)
```
